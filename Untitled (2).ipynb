```python
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
X=df_copy.drop('FWI',axis=1)
```

```python
Y=df_copy['FWI']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
X_train,X_test
```

```python
X_train.corr()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_trained=scaler.fit_transform(X_train)
```

```python
X_tested=scaler.transform(X_test)
```

```python
X_trained.shape,X_tested.shape,Y_train.shape
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
from sklearn.linear_model import Ridge
```

```python
lasso=Lasso()
```

```python
lasso.fit(X_trained,Y_train)
```

```python
Y_pred=lasso.predict(X_tested)
```

```python
print(r2_score(Y_test,Y_pred))
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
```

```python
ridge.fit(X_trained,Y_train)
```

```python
Y_predr=ridge.predict(X_tested)
```

```python
print(r2_score(Y_test,Y_predr))
```
